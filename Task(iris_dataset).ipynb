3) Logistic Regression 

```python
from sklearn.datasets import load_iris,fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#Logistic Regression
df=load_iris() #load the data
x_train,x_test,y_train,y_test=train_test_split(df.data,df.target,test_size=0.3,random_state=42) # training and testing of model
```

```python
model_iris=LogisticRegression(max_iter=200) #iteration process
```

```python
model_iris.fit(x_train,y_train) # fit the model
```

```python
y_pred=model_iris.predict(x_test) #predit the model
```

3) Accuracy of Model

```python
accuracy=accuracy_score(y_test,y_pred) # accuracy of model
print(accuracy)
```
